```python
def get_company_information(domain: str) -> (dict, bool):
    response = requests.request(
    "GET",
    f"https://company.bigpicture.io/v1/companies/find?domain={domain}",
    headers={"Authorization": BIGPICTURE_AUTH_KEY})
    # check if key is valid
    if response.status_code == 401:
        print("no valid api key was supplied")
        return None, False
    # check if domain is in database
    if response.status_code == 404:
        return None, False
    # check if data is stale and has to be queried for again later
    if response.status_code == 202:
        #return None, True
        return response.json()
    # data is fresh and got delivered to us -> return payload
    return response.json()
#  return response.json(), False
```

```python
def read_json(filename: str) -> dict:

    try:
        with open(filename, "r") as f:
            data = json.loads(f.read())
    except:
        raise Exception(f"Reading {filename} file encountered an error")

    return data
def normalize_json(data: dict) -> dict:

    new_data = dict()
    for key, value in data.items():
        if not isinstance(value, dict):
            new_data[key] = value
        else:
            for k, v in value.items():
                new_data[key + "_" + k] = v

    return new_data

def generate_csv_data(data: dict) -> str:

    csv_columns = data.keys()


    csv_data = ",".join(csv_columns) + "\n"

    new_row = list()
    for col in csv_columns:
        new_row.append(str(data[col]))


    csv_data += ",".join(new_row) + "\n"

    return csv_data
```
